## DATA CLEANING

```python
df.rename(columns={'User_ID':'user_id', 'Cust_name':'cust_name', 'Product_ID':'product_id', 'Gender':'gender', 'Age Group':'age group', 'Age':'age',
       'Marital_Status':'marital_status', 'State':'state', 'Zone':'zone', 'Occupation':'job', 'Product_Category':'product_category',
       'Orders':'order', 'Amount':'amount', 'Status':'status', 'unnamed1':'unnamed1'},inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels = False)
```

```python
df.columns
```

```python
##delet empty columns
df.drop(['status','unnamed1'],axis=1,inplace=True)
```

```python
## delet null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
## data type chenge
df['amount']=df['amount'].astype('int')
```

```python
df['amount'].dtypes
```

-->delete null valuse

-->delete empty columns

-->chenge data type of amount

```python
df['gender'].value_counts()
```

```python
sns.countplot(x='gender',data=df,hue='gender')
plt.legend(['male','female'])
plt.show()
```

```python
plt.pie(df['gender'].value_counts(),labels=df['gender'].value_counts().index,autopct='%1.2f%%')
plt.title('percentage in each gender')
plt.show()
```

```python
df.columns
```

```python
#top state most of order
top_states = df.groupby('state')['order'].sum().sort_values(ascending=False).head(10)
```

```python
top_states.plot(kind='bar')
plt.title('Top 10 most of order')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
#top 10 state by total amount
top_states_amount = df.groupby('state')['amount'].sum().sort_values(ascending=False).head(10)
```

```python
top_states_amount.plot(kind='bar', color='green')
plt.title('Top 10 States by Total Amount')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

## observation
-->highest of the orders are uttar pradesh,maharastra,karnataka

-->lowest of the order are gujurat,haryana,kerala

-->highest sales amount is from uttar pradesh,karnataka,maharastra

-->lowes sales amount is from gujurat,bihar, hariyana
